# Dashboard - Earned Run Average

```python
pitcher_names = ['Alexander, Scott', 'Baez, Pedro', 'Buehler, Walker', 'Ferguson, Caleb', 'Floro, Dylan',
                 'Gonsolin, Tony', 'Gonzalez, Victor', 'Graterol, Brusdar', 'Jansen, Kenley', 'Kelly, Joe',
                 'Kershaw, Clayton', 'Kolarek, Adam', 'May, Dustin', 'McGee, Jake', 'Santana, Dennis', 
                 'Sborz, Josh', 'Stripling, Ross', 'Treinen, Blake', 'Urias, Julio', 'White, Mitch', 'Wood, Alex']
```

```python
import plotly.express as px

def pitcher_ERA(pitcher_name, games):
    
    fig = px.line(x=list(range(games)), y=4*np.ones(games))
    
    try:
        df2018_pitch = game_logs_2018.loc[pitcher_name]
        fig.add_scatter(x=list(range(len(df2018_pitch.Date))), y=df2018_pitch.ERA, name='2018')
        
    except:
        pass
    
    try:
        df2019_pitch = game_logs_2019.loc[pitcher_name]
        fig.add_scatter(x=list(range(len(df2019_pitch.Date))), y=df2019_pitch.ERA, name='2019')
    
    except:
        pass
    
    try:
        df2020_pitch = game_logs_2020.loc[pitcher_name]
        fig.add_scatter(x=list(range(len(df2020_pitch.Date))), y=df2020_pitch.ERA, name='2020')
    
    except:
        pass
    
    fig.update_layout(title='Earned Run Average - ' + pitcher_name, xaxis_title="Game", yaxis_title="ERA", 
                      legend_title="Season")
                     
    fig.show()
```

```python
pitcher_ERA('Kershaw, Clayton', 12)
```

```python
import numpy as np
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
df = px.data.tips()
# Build App
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Earned Run Average"),
    #dcc.Graph(id='graph'),
    html.Label([
        "Player",
        dcc.Dropdown(
            id='player-dropdown', clearable=False,
            value='Alexander, Scott', options=[
                {'label': name, 'value': name}
                for name in pitcher_names
            ])
    ]), dcc.Graph(id='graph')
])
# Define callback to update graph
@app.callback(
    Output('graph', 'figure'),
    [Input("player-dropdown", "value")]
)
def update_figure(pitcher_name):
    return pitcher_ERA(pitcher_name)
# Run app and display result inline in the notebook
app.run_server(mode='inline')
#Dropdown names have to match as well as graph id's.
#"Value" in input(Line 25) will be what goes into function (Line 27, "colorscale")
#List of player names will go into Lines 17 & 18 - label and value will be the same. Define list into cell above
```

```python
def pitcher_WHIP(pitcher_name, games):
    
    fig = px.line(x=list(range(games)), y=1.3*np.ones(games))
    
    try:
        df2018_pitch_WHIP = game_logs_2018.loc[pitcher_name]
        fig.add_scatter(x=list(range(len(df2018_pitch_WHIP.Date))), y=df2018_pitch_WHIP.WHIP, name='2018')
        
    except:
        pass
    
    try:
        df2019_pitch_WHIP = game_logs_2019.loc[pitcher_name]
        fig.add_scatter(x=list(range(len(df2019_pitch_WHIP.Date))), y=df2019_pitch_WHIP.WHIP, name='2019')
    
    except:
        pass
    
    try:
        df2020_pitch_WHIP = game_logs_2020.loc[pitcher_name]
        fig.add_scatter(x=list(range(len(df2020_pitch_WHIP.Date))), y=df2020_pitch_WHIP.WHIP, name='2020')
    
    except:
        pass
    
    fig.update_layout(title='Walks & Hits Per Inning Pitched - ' + pitcher_name, 
                      xaxis_title="Game", yaxis_title="WHIP", legend_title="Season")
    fig.show()
```

```python
def position_SLG(player_name, games):
    
    fig = px.line(x=list(range(games)), y=.410*np.ones(games))
    
    try:
        df2018_position_SLG = game_logs_2018_position.loc[player_name]
        fig.add_scatter(x=list(range(len(df2018_position_SLG.Date))), y=df2018_position_SLG.SLG, name='2018')
        
    except:
        pass
    
    try:
        df2019_position_SLG = game_logs_2019_position.loc[player_name]
        fig.add_scatter(x=list(range(len(df2019_position_SLG.Date))), y=df2019_position_SLG.SLG, name='2019')
    
    except:
        pass
    
    try:
        df2020_position_SLG = game_logs_2020_position.loc[player_name]
        fig.add_scatter(x=list(range(len(df2020_position_SLG.Date))), y=df2020_position_SLG.SLG, name='2020')
    
    except:
        pass
    
    fig.update_layout(title='Slugging Percentage - ' + player_name, xaxis_title="Game", yaxis_title="SLG", 
                      legend_title="Season")
                     
    fig.show()
```
